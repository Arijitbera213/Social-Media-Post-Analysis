```python
for i in USER_STATE:
    translations[i] = translator.translate(i).text
print(translations)
first1.replace({"USER_STATE": translations},inplace=True)
first1
first1.to_csv('First.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
age1_10 = data2.POST_ID_Liked_by_user.value_counts()[(data2.POST_ID_Liked_by_user.value_counts() <= 10) & (data2.POST_ID_Liked_by_user.value_counts() >= 1)]
age10_20 = data2.POST_ID_Liked_by_user.value_counts()[(data2.POST_ID_Liked_by_user.value_counts() <= 20) & (data2.POST_ID_Liked_by_user.value_counts() >= 10)]
age20_30 = data2.POST_ID_Liked_by_user.value_counts()[(data2.POST_ID_Liked_by_user.value_counts() <= 30) & (data2.POST_ID_Liked_by_user.value_counts() >= 20)]
age30_40 = data2.POST_ID_Liked_by_user.value_counts()[(data2.POST_ID_Liked_by_user.value_counts() <= 40) & (data2.POST_ID_Liked_by_user.value_counts() >= 30)]
age40above = data2.POST_ID_Liked_by_user.value_counts()[data2.POST_ID_Liked_by_user.value_counts() >= 40]

x = ["1-10","10-20","20-30","30-40","40+"]
y = [len(age1_10.values),len(age10_20.values),len(age20_30.values),len(age30_40.values),len(age40above.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=x, y=y, palette="rocket")
plt.title("Number of LIkes")
plt.xlabel("Likes")
plt.ylabel("Post")
plt.show()
```
